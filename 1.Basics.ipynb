## Tensors

### Simple Two-layer Network

#### Numpy version

```python
import numpy as np
```

```python
# batch_size
N = 64
# input dimension
D_in = 1000
# hidden dimension
H = 100
# output dimension
D_out = 10
```

```python
# random input and output data
x = np.random.randn(N, D_in)
y = np.random.randn(N, D_out)
```

```python
# randomly initialize weights
w1 = np.random.randn(D_in, H)
w2 = np.random.randn(H, D_out)
```

```python
# learning rate
l_rate = 1e-6
```

```python
for t in range(500):
    # Forward pass
    h = x.dot(w1)
    h_relu = np.maximum(h, 0)
    y_pred = h_relu.dot(w2)
    
    # compute and print loss
    loss = np.square(y_pred - y).sum()
    print(t, loss)
    
    # Backprop to compute gradients
    dy_pred = 2.0 * (y_pred - y)
    #dw2/dloss
    dw2 = h_relu.T.dot(dy_pred)
    
    dh_relu = dy_pred.dot(w2.T)
    
    dh = dh_relu.copy()
    dh[h < 0] = 0
    
    #dw1/dloss
    dw1 = x.T.dot(dh)
    
    # update weights 
    w1 -= l_rate * dw1
    w2 -= l_rate * dw2
```

#### PyTorch version

```python
import torch
```

```python
# run on GPU
dtype = torch.cuda.FloatTensor
```

```python
x = torch.randn(N, D_in).type(dtype)
y = torch.randn(N, D_out).type(dtype)
```

```python
w1 = torch.randn(D_in, H).type(dtype)
w2 = torch.randn(H, D_out).type(dtype)
```

```python
for t in range(500):
    h = x.mm(w1)
    h_relu = h.clamp(min=0)
    y_pred = h_relu.mm(w2)
    
    loss = (y_pred - y).pow(2).sum()
    print(t, loss)
    
    dy_pred = 2.0 * (y_pred - y)
    dw2 = h_relu.t().mm(dy_pred)
    dh_relu = dy_pred.mm(w2.t())
    dh = dh_relu.clone()
    dh[h < 0] = 0
    dw1 = x.t().mm(dh)
    
    w1 -= l_rate * dw1
    w2 -= l_rate * dw2
```

## Autograd

### Variables and autograd

```python
from torch.autograd import Variable
```

```python
x = Variable(torch.randn(N, D_in).type(dtype), requires_grad=False)
y = Variable(torch.randn(N, D_out).type(dtype), requires_grad=False)
```

```python
w1 = Variable(torch.randn(D_in, H).type(dtype), requires_grad=True)
w2 = Variable(torch.randn(H, D_out).type(dtype), requires_grad=True)
```

```python
for t in range(500):
    y_pred = x.mm(w1).clamp(min=0).mm(w2)
    
    loss = (y_pred - y).pow(2).sum()
    print(t, loss.data[0])
    
    loss.backward()
    
    w1.data -= l_rate * w1.grad.data
    w2.data -= l_rate * w2.grad.data
    
    # manually zero gradients after updating weights
    w1.grad.data.zero_()
    w2.grad.data.zero_()
```

### Custom autograd functions

```python
class MyReLU(torch.autograd.Function):
    def forward(self, input):
        self.save_for_backward(input)
        return input.clamp(min=0)
    
    def backward(self, grad_output):
        inp, = self.saved_tensors
        grad_input = grad_output.clone()
        grad_input[inp < 0] = 0
        return grad_input
```

```python
for t in range(500):
    relu = MyReLU()
    
    y_pred = relu(x.mm(w1)).mm(w2)
    
    loss = (y_pred - y).pow(2).sum()
    print(t, loss.data[0])
    
    loss.backward()
    
    w1.data -= l_rate * w1.grad.data
    w2.data -= l_rate * w2.grad.data
    
    w1.grad.data.zero_()
    w2.grad.data.zero_()
```

## Pytorch modules

### nn

```python
x = Variable(torch.randn(N, D_in))
y = Variable(torch.randn(N, D_out), requires_grad=False)
l_rate = 1e-4
```

```python
model = torch.nn.Sequential(
    torch.nn.Linear(D_in, H),
    torch.nn.ReLU(),
    torch.nn.Linear(H, D_out)
)

loss_fn = torch.nn.MSELoss(size_average=False)
```

```python
for t in range(500):
    y_pred = model(x)
    loss = loss_fn(y_pred, y)
    print(t, loss.data[0])
    
    # zero gradients before running backprop
    model.zero_grad()
    
    loss.backward()
    
    #update weights using gradient descent
    for param in model.parameters():
        param.data -= l_rate * param.grad.data
```

### optim

```python
optimizer = torch.optim.Adam(model.parameters(), lr=l_rate)
```

```python
for t in range(500):
    y_pred = model(x)
    loss = loss_fn(y_pred, y)
    print(t, loss.data[0])
    
    optimizer.zero_grad()
    
    loss.backward()
    
    optimizer.step()
```

### Custom nn modules

```python
class TwoLayerNet(torch.nn.Module):
    def __init__(self, D_in, H, D_out):
        super(TwoLayerNet, self).__init__()
        self.linear1 = torch.nn.Linear(D_in, H)
        self.linear2 = torch.nn.Linear(H, D_out)
    
    def forward(self, x):
        h_relu = self.linear1(x).clamp(min=0)
        y_pred = self.linear2(h_relu)
        return y_pred
```

```python
model = TwoLayerNet(D_in, H, D_out)
```

```python
loss_fn = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(), lr=1e-4)
```

```python
for t in range(500):
    y_pred = model(x)
    loss = loss_fn(y_pred, y)
    print(t, loss.data[0])
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

## Control FLow + Weight Sharing

```python
import random
```

```python
class DynamicNet(torch.nn.Module):
    def __init__(self, D_in, H, D_out):
        super(DynamicNet, self).__init__()
        self.input_linear = torch.nn.Linear(D_in, H)
        self.middle_linear = torch.nn.Linear(H, H)
        self.output_linear = torch.nn.Linear(H, D_out)
    
    def forward(self, x):
        h_relu = self.input_linear(x).clamp(min=0)
        
        for _ in range(random.randint(0, 3)):
            h_relu = self.middle_linear(h_relu).clamp(min=0)
        
        y_pred = self.output_linear(h_relu)
        return y_pred
```

```python
model = DynamicNet(D_in, H, D_out)
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=1e-4, momentum=0.9)
```

```python
for t in range(500):
    y_pred = model(x)
    loss = loss_fn(y_pred, y)
    print(t, loss.data[0])
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
